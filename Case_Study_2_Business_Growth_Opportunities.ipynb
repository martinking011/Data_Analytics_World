# 1. Monthly Revenue Trend Analysis by Product

This section analyzes how monthly revenue varies across FinTechLend's three main lending products: **Business Loan**, **Personal Loan**, and **BNPL**.

Understanding revenue momentum at the product level helps identify:
- Seasonal patterns or revenue peaks,
- Product performance consistency,
- Opportunities for prioritizing product-specific growth strategies.

We use Plotly to generate an interactive line chart with monthly data aggregated per product.


```python
# Step 1: Convert date and prepare data
business_performance['date'] = pd.to_datetime(business_performance['date'])

# Step 2: Aggregate monthly by product
monthly_perf = business_performance.groupby(
    [pd.Grouper(key='date', freq='ME'), 'product']
).agg({
    'application_volume': 'sum',
    'conversion_rate': 'mean',
    'acquisition_cost': 'mean',
    'revenue': 'sum',
    'margin': 'mean'
}).reset_index()

# Format date for display
monthly_perf['month_label'] = monthly_perf['date'].dt.strftime('%b-%y')
monthly_perf = monthly_perf.sort_values(by='date')

# Step 3: Plot using Plotly Express
fig = px.line(
    monthly_perf,
    x='month_label',
    y='revenue',
    color='product',
    markers=True,
    title='Monthly Revenue Trend by Product',
    labels={'month_label': 'Month', 'revenue': 'Revenue (₦)', 'product': 'Product'},
    color_discrete_sequence=px.colors.qualitative.Set1
)

# Step 4: Layout styling
fig.update_layout(
    title_font=dict(size=20),
    font=dict(size=12),
    xaxis_tickangle=-45,
    plot_bgcolor='white',
    paper_bgcolor='white',
    margin=dict(t=80, l=40, r=20, b=40),
    hovermode='x unified',
    legend_title_text='Product'
)

fig.show()
```

## Detailed Insight – Monthly Revenue Trend by Product

This visualization captures monthly revenue trends across FinTechLend's three key lending products — Business Loans, Personal Loans, and BNPL — over the observed period. It allows us to assess relative product performance, seasonality, and directional momentum.

---

### I. Business Loan – Consistently Strong Performer
- **Revenue Leadership**: Business Loans consistently contribute the highest monthly revenue, reflecting strong demand and higher disbursement values.
- **Profitability Support**: This product also exhibits the highest net interest margin (10%), confirming that volume is backed by strong unit economics.
- **Strategic Implication**: Business Loans should remain the flagship offering, with continued investment in risk-based pricing and acquisition.

---

### II. Personal Loan – Stable and Reliable Growth
- **Steady Trajectory**: Revenue from Personal Loans demonstrates consistent month-over-month performance, with limited volatility.
- **Demographic Fit**: Strong interest among midlife professionals (36–55 age group) supports this product’s ongoing relevance.
- **Strategic Implication**: Position Personal Loans as a mainstream credit option for salaried professionals and increase cross-sell into this base.

---

### III. BNPL – Volatile but Popular
- **Revenue Fluctuations**: BNPL revenue shows sharp month-to-month variability, peaking during certain quarters and dipping unpredictably.
- **Risk Exposure**: Despite competitive ARPU, BNPL yields the **lowest margin (6%)** and **highest loss rate (94%)**, signaling credit risk or operational friction.
- **Strategic Implication**: BNPL requires either redesign (example:- tighter repayment terms or better risk segmentation) or careful de-prioritization.

---

### General Trends & Seasonality
- **March Surges**: Revenue spikes in March for all products, possibly linked to quarterly disbursement cycles, salary flows, or targeted campaigns.
- **Mid-Year Dips**: Revenue declines during July–September, suggesting seasonal softening — a potential opportunity for marketing activation.

### Strategic Recommendations

# 2. Channel Efficiency Analysis: Acquisition Cost vs Revenue

This section evaluates the performance of FinTechLend's acquisition channels based on three metrics:
- **Average Acquisition Cost (CAC)**  
- **Total Revenue Generated**  
- **Conversion Rate (%)**

We use a **bubble chart** to assess the relative efficiency of each channel — combining cost, return, and conversion size into one visual.

This helps identify high-ROI channels and flag inefficient or overpriced acquisition sources.


```python
# Helper function to shorten large currency values
def format_naira(value):
    if value >= 1e12:
        return f"₦{value / 1e12:.1f}T"
    elif value >= 1e9:
        return f"₦{value / 1e9:.1f}B"
    elif value >= 1e6:
        return f"₦{value / 1e6:.0f}M"
    else:
        return f"₦{value:,.0f}"

# Step 1: Group and aggregate
channel_perf = business_performance.groupby('channel').agg({
    'application_volume': 'sum',
    'conversion_rate': 'mean',
    'acquisition_cost': 'mean',
    'revenue': 'sum',
    'margin': 'mean'
}).reset_index()

channel_perf = channel_perf.sort_values(by='conversion_rate', ascending=False)

# Step 2: Base scatter plot
fig = px.scatter(
    channel_perf,
    x='acquisition_cost',
    y='revenue',
    size='conversion_rate',
    color='channel',
    hover_name='channel',
    size_max=60,
    title='Channel Efficiency: Acquisition Cost vs Revenue',
    labels={
        'acquisition_cost': 'Avg Acquisition Cost (₦)',
        'revenue': 'Total Revenue (₦)',
        'conversion_rate': 'Conversion Rate',
        'channel': 'Channel'
    },
    color_discrete_sequence=px.colors.qualitative.Set1
)

# Step 3: Add dark black text labels to each bubble
for i, row in channel_perf.iterrows():
    label = f"{row['channel']}<br>Cost: {format_naira(row['acquisition_cost'])}<br>Rev: {format_naira(row['revenue'])}"

    fig.add_annotation(
        x=row['acquisition_cost'],
        y=row['revenue'],
        text=label,
        showarrow=False,
        font=dict(color='black', size=11),
        bgcolor='rgba(255,255,255,0.6)',
        xanchor='center',
        yanchor='middle'
    )

# Step 4: Layout tweaks
fig.update_layout(
    title_font=dict(size=20),
    font=dict(size=12),
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='closest',
    margin=dict(t=80, l=40, r=20, b=40),
    legend_title_text='Channel'
)

fig.show()
```

## Detailed Insight – Channel Efficiency (CAC vs Revenue)

---

### I. Partner Channel – High Revenue, High Cost
- Generates the **highest total revenue** (₦116M) and leads in conversion rate (8%).
- However, it comes at a premium, with the **highest acquisition cost (₦300)**.
- Best suited for high-intent, high-value acquisition bursts — not ideal for always-on scaling.

---

### II. Direct Channel – Most Efficient
- Achieves a strong **revenue output (₦72M)** at the **lowest CAC (₦50)** with a solid 5% conversion rate.
- Represents the **most scalable and cost-efficient channel**.
- Should be prioritized in both paid and organic growth campaigns.

---

### III. Content & Social Media – Low ROI
- Both channels show **low revenue (₦29M–₦43M)** despite moderate acquisition costs (₦100–₦150).
- Conversion rates are the **lowest (2–3%)**, indicating a misalignment between message and market.
- These should be **re-evaluated or repositioned** for top-of-funnel rather than direct acquisition.

---

### Strategic Takeaways
- **Double down on Direct** for efficient scaling.
- **Use Partner for spikes**, but manage CAC carefully.
- **Audit low-performing channels** (example:- Social and Content) for creative, audience, or channel mismatch.
- Consider introducing **multi-touch attribution** to better value early-stage influence channels.



# 3. Customer Sentiment by Segment – Satisfaction vs NPS

Understanding how different customer segments perceive FinTechLend's service is crucial to reducing churn and improving loyalty. This analysis maps each customer segment by:

- **Satisfaction Score (1–5 scale)**
- **Net Promoter Score (NPS)**
- **Feedback Volume**

We use a **bubble chart** to highlight high-impact segments that may require targeted retention, better onboarding, or tailored support improvements.


```python
# Convert date column to datetime
customer_feedback['date'] = pd.to_datetime(customer_feedback['date'])

# Aggregate satisfaction and NPS by customer segment
segment_feedback = customer_feedback.groupby('customer_segment').agg({
    'nps_score': 'mean',
    'satisfaction_score': 'mean',
    'feedback_volume': 'sum'
}).reset_index()

# Format feedback volume
def format_volume(v):
    if v >= 1e6:
        return f"{v / 1e6:.1f}M"
    elif v >= 1e3:
        return f"{v / 1e3:.0f}K"
    else:
        return str(int(v))

# Base scatter plot
fig = px.scatter(
    segment_feedback,
    x='satisfaction_score',
    y='nps_score',
    size='feedback_volume',
    color='customer_segment',
    hover_name='customer_segment',
    size_max=60,
    title='Customer Sentiment by Segment: Satisfaction vs NPS',
    labels={
        'satisfaction_score': 'Avg Satisfaction Score',
        'nps_score': 'Avg NPS Score',
        'feedback_volume': 'Feedback Volume',
        'customer_segment': 'Customer Segment'
    },
    # Changed 'Tab10' to 'Plotly' which is a valid qualitative color sequence in Plotly Express
    color_discrete_sequence=px.colors.qualitative.Plotly
)

# Add black labels inside the bubbles
for i, row in segment_feedback.iterrows():
    label = f"{row['customer_segment']}<br>S:{row['satisfaction_score']:.1f} N:{row['nps_score']:.1f}<br>F:{format_volume(row['feedback_volume'])}"

    fig.add_annotation(
        x=row['satisfaction_score'],
        y=row['nps_score'],
        text=label,
        showarrow=False,
        font=dict(color='black', size=11),
        bgcolor='rgba(255,255,255,0.6)',
        xanchor='center',
        yanchor='middle'
    )

# Layout styling
fig.update_layout(
    title_font=dict(size=18),
    font=dict(size=12),
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='closest',
    margin=dict(t=80, l=40, r=20, b=40),
    legend_title_text='Customer Segment'
)

fig.show()
```

## Detailed Insight – Customer Sentiment by Segment

---

### I. High-Value Advocates (Gen Z, Small Business)
- These segments show **strong satisfaction (4.5+) and high NPS (>45)**.
- Feedback volume is moderate but influential, suggesting positive user experience and referral potential.
- Opportunity: Use these groups to drive word-of-mouth campaigns and referral incentives.

---

### II. At-Risk Segments (Mid-Career, Startups)
- These segments have **lower-than-average NPS (44.5)** and moderate satisfaction (4.3–4.4).
- Feedback volumes are high, implying **engaged but dissatisfied users**.
- Priority for churn prevention and service recovery efforts.

---

### III. Stable but Passive (Gen X, Medium Enterprises)
- These users sit near the average on both axes (Satisfaction 4.4, NPS 45.0).
- Their experience is neither negative nor enthusiastic — a potential **“silent churn”** group.
- Opportunity: Re-engage through onboarding refresh, surveys, or proactive support.

---

### Strategic Takeaways
- **Protect and amplify your promoters** (Gen Z, Small Business) through loyalty, recognition, and referral initiatives.
- **Fix friction points** for Mid-Career and Startup users with faster feedback loops and targeted CX improvements.
- **Reinvigorate the middle** — Gen X and Medium Enterprises — with personalized journeys and engagement nudges.

# 5. Market Size vs Competitor Revenue Over Time

This analysis tracks the evolution of the total addressable market and competitor revenue in the fintech lending space. It helps us understand:

- How fast the market is growing (₦ trillions per month)
- Whether competitors are scaling proportionately
- Where FinTechLend sits in this competitive trajectory

The dual-line chart compares **Market Size** and **Competitor Revenue** monthly, offering a clear view of scale and share shifts.


```python
# Format large numbers (₦ Millions, Billions, Trillions)
def format_naira(value):
    if value >= 1e12:
        return f"₦{value / 1e12:.1f}T"
    elif value >= 1e9:
        return f"₦{value / 1e9:.0f}B"
    elif value >= 1e6:
        return f"₦{value / 1e6:.0f}M"
    else:
        return f"₦{value:,.0f}"

# Clean and prepare data
market_data.columns = market_data.columns.str.strip()
market_data['market_size'] = market_data['market_size'].replace('[\$,]', '', regex=True).astype(float)
market_data['competitor_revenue'] = market_data['competitor_revenue'].replace('[\$,]', '', regex=True).astype(float)
market_data['date'] = pd.to_datetime(market_data['date'])

# Remove duplicate market_size entries by keeping only one per product per month
unique_market_sizes = market_data.drop_duplicates(subset=['date', 'product'])[['date', 'market_size']]
monthly_market_size = unique_market_sizes.groupby(pd.Grouper(key='date', freq='ME'))['market_size'].sum().reset_index()

# Compute monthly competitor revenue
monthly_revenue = market_data.groupby(pd.Grouper(key='date', freq='ME'))['competitor_revenue'].sum().reset_index()

# Merge market size and revenue
market_trend = pd.merge(monthly_market_size, monthly_revenue, on='date')
market_trend['month_label'] = market_trend['date'].dt.strftime('%b-%y')

# Create figure
fig = go.Figure()

# Line 1: Market Size
fig.add_trace(go.Scatter(
    x=market_trend['month_label'],
    y=market_trend['market_size'],
    name='Market Size (₦)',
    mode='lines+markers',
    line=dict(color='steelblue', width=3),
    marker=dict(symbol='circle')
))

# Line 2: Competitor Revenue (right axis)
fig.add_trace(go.Scatter(
    x=market_trend['month_label'],
    y=market_trend['competitor_revenue'],
    name='Competitor Revenue (₦)',
    mode='lines+markers',
    line=dict(color='firebrick', width=3),
    marker=dict(symbol='square'),
    yaxis='y2'
))

# Add clean black labels
for i, row in market_trend.iterrows():
    fig.add_annotation(
        x=row['month_label'],
        y=row['market_size'],
        text=format_naira(row['market_size']),
        showarrow=False,
        font=dict(color='black', size=11),
        bgcolor='rgba(255,255,255,0.6)',
        yanchor='bottom'
    )

    fig.add_annotation(
        x=row['month_label'],
        y=row['competitor_revenue'],
        text=format_naira(row['competitor_revenue']),
        showarrow=False,
        font=dict(color='black', size=11),
        bgcolor='rgba(255,255,255,0.6)',
        yanchor='top',
        yref='y2'
    )

# Layout
fig.update_layout(
    title='Market Size vs Competitor Revenue Over Time',
    title_font=dict(size=20),
    xaxis=dict(title='Month', tickangle=-45, showgrid=False),
    yaxis=dict(title='Market Size (₦)', titlefont=dict(color='steelblue'), tickfont=dict(color='steelblue')),
    yaxis2=dict(
        title='Competitor Revenue (₦)',
        titlefont=dict(color='firebrick'),
        tickfont=dict(color='firebrick'),
        anchor='x',
        overlaying='y',
        side='right'
    ),
    legend=dict(
        title='Metrics',
        orientation='v',
        x=1.02,
        y=1,
        xanchor='left',
        bgcolor='rgba(255,255,255,0.95)',
        bordercolor='lightgray',
        borderwidth=1
    ),
    margin=dict(t=80, l=50, r=100, b=60),
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='x unified',
    font=dict(size=12)
)

fig.show()
```

```python
# Rebuild the entire process to ensure all variables are defined

# Format helpers
def format_naira(val):
    if val >= 1e12: return f"₦{val / 1e12:.2f}T"
    elif val >= 1e9: return f"₦{val / 1e9:.2f}B"
    elif val >= 1e6: return f"₦{val / 1e6:.0f}M"
    else: return f"₦{val:,.0f}"

def format_precise_percent(val):
    return f"{val * 100:.7f}%" if pd.notnull(val) else ""

def format_arrow_color(change):
    if pd.isnull(change): return "black"
    return "green" if change > 0 else "red"

def merge_arrow_into_share(curr, change):
    base = f"{curr * 100:.7f}%"
    if pd.isnull(change):
        return base
    elif change > 0:
        return f"{base} ↑"
    elif change < 0:
        return f"{base} ↓"
    else:
        return base

# Preprocess original data
business_performance['date'] = pd.to_datetime(business_performance['date'])
market_data['date'] = pd.to_datetime(market_data['date'])
market_data['market_size'] = market_data['market_size'].replace('[\$,]', '', regex=True).astype(float)
market_data['competitor_revenue'] = market_data['competitor_revenue'].replace('[\$,]', '', regex=True).astype(float)

# Aggregation
fintechlend_monthly = business_performance.groupby(pd.Grouper(key='date', freq='ME')).agg({
    'revenue': 'sum'
}).rename(columns={'revenue': 'FinTechLend_Revenue'}).reset_index()

market_monthly = market_data.groupby(pd.Grouper(key='date', freq='ME')).agg({
    'market_size': 'sum',
    'competitor_revenue': 'sum'
}).reset_index()

# Merge
merged = pd.merge(fintechlend_monthly, market_monthly, on='date', how='left')
merged['Total_Market_Size'] = merged['FinTechLend_Revenue'] + merged['competitor_revenue']
merged['FinTechLend_Market_Share'] = merged['FinTechLend_Revenue'] / merged['Total_Market_Size']
merged['Market_Share_Change'] = merged['FinTechLend_Market_Share'].diff()

# Formatting
merged['Month'] = merged['date'].dt.strftime('%b-%Y')
merged['Revenue'] = merged['FinTechLend_Revenue'].apply(format_naira)
merged['Market Size'] = merged['Total_Market_Size'].apply(format_naira)
merged['Market Share with Arrows'] = merged.apply(
    lambda row: merge_arrow_into_share(row['FinTechLend_Market_Share'], row['Market_Share_Change']),
    axis=1
)
merged['Arrow Color'] = merged['Market_Share_Change'].apply(format_arrow_color)

# Table layout
row_height = 30
table_height = 140 + (len(merged) * row_height)

# Plot the final table with arrows embedded in Market Share column
fig = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>Month</b>", "<b>Revenue</b>", "<b>Market Size</b>", "<b>Market Share</b>"],
        fill_color='#003366',
        font=dict(color='white', size=13),
        align='left'
    ),
    cells=dict(
        values=[
            merged['Month'],
            merged['Revenue'],
            merged['Market Size'],
            merged['Market Share with Arrows']
        ],
        font=dict(size=12),
        align='left',
        fill_color=[['#f7f7f7', '#ffffff'] * (len(merged) // 2 + 1)],
        line_color='lightgray',
        height=row_height,
        font_color=[
            "black",  # Month
            "black",  # Revenue
            "black",  # Market Size
            merged['Arrow Color']  # Market Share
        ]
    )
)])

fig.update_layout(
    title='FinTechLend Monthly Revenue & Market Share',
    title_font=dict(size=20),
    margin=dict(l=20, r=20, t=80, b=20),
    height=table_height
)

fig.show()

```

# 17. Group Performance Test – Conversion Rate by Channel

This test checks whether **conversion rates differ significantly across acquisition channels** using two statistical methods:

- **One-Way ANOVA** – assumes normal distribution and equal variances  
- **Kruskal-Wallis H Test** – a non-parametric alternative, does not assume normality

These methods help confirm whether observed differences in conversion rate are **statistically significant** or due to random variation.


```python
from scipy.stats import f_oneway, kruskal

# Group conversion rates by channel
channel_groups = [
    group['conversion_rate'].values
    for _, group in business_performance.groupby('channel')
]

# Perform ANOVA and Kruskal-Wallis tests
anova_result = f_oneway(*channel_groups)
kruskal_result = kruskal(*channel_groups)

# Prepare summary dictionary
anova_summary = {
    "Metric": ["ANOVA F-statistic", "ANOVA p-value", "Kruskal-Wallis H-statistic", "Kruskal-Wallis p-value"],
    "Value": [
        round(anova_result.statistic, 4),
        round(anova_result.pvalue, 4),
        round(kruskal_result.statistic, 4),
        round(kruskal_result.pvalue, 4)
    ]
}

# Create Plotly table
fig = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>Metric</b>", "<b>Value</b>"],
        fill_color='#003366',
        font=dict(color='white', size=14),
        align='left'
    ),
    cells=dict(
        values=[anova_summary["Metric"], anova_summary["Value"]],
        fill_color='lavender',
        align='left',
        font=dict(size=13)
    )
)])

fig.update_layout(
    title='Group Performance Statistical Test Summary',
    title_font=dict(size=20),
    height=350,
    margin=dict(l=20, r=20, t=80, b=20)
)

fig.show()
```

## Detailed Insight – Are Channel Conversion Rates Statistically Different?

---

### ANOVA Result
- **F-statistic** indicates how much variance exists between group means relative to within-group variance.
- A **p-value < 0.05** would mean **at least one channel** differs significantly in conversion rate.
- If p > 0.05, we fail to reject the null hypothesis that all means are equal.

---

### Kruskal-Wallis Result
- Non-parametric alternative to ANOVA (useful when distributions aren't normal).
- Similar interpretation: **low p-value means significant differences** between at least one pair of groups.

---

### Interpretation Guidelines

- If **both p-values < 0.05** → Significant differences exist across channels. Proceed with post-hoc tests (e.g., Tukey HSD or Dunn's test).
- If **both p-values ≥ 0.05** → No statistically significant difference in conversion rate by channel. Focus on operational improvements.

---

### Strategic Implication

- **Statistically significant results** justify **channel-specific optimization** or investment.
- If not significant, observed conversion differences may not be meaningful — prioritize **cost efficiency and volume** instead.

> This test ensures data-driven decisions when comparing performance across acquisition channels.

# 19. Time Series Decomposition of Monthly Revenue

This analysis applies an **additive decomposition** to monthly revenue using a **6-month seasonality window**. The goal is to isolate:

- **Trend** – long-term direction of revenue  
- **Seasonality** – repeated cyclical patterns (example:- biannual spikes)  
- **Residuals** – unexplained variation or noise

This breakdown helps diagnose whether revenue changes are structural, cyclical, or random.


```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 1: Prepare and aggregate monthly revenue
ts_data = business_performance.copy()
ts_data['date'] = pd.to_datetime(ts_data['date'])

monthly_revenue = ts_data.groupby(pd.Grouper(key='date', freq='ME'))['revenue'].sum()

# Step 2: Decompose the time series with 6-month periodicity
decomposition = seasonal_decompose(monthly_revenue, model='additive', period=6)

# Step 3: Prepare subplot structure
fig = make_subplots(
    rows=4, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.06,
    subplot_titles=[
        "Observed Revenue",
        "Trend Component",
        "Seasonal Component",
        "Residuals"
    ]
)

# Step 4: Add each component as a line trace
fig.add_trace(go.Scatter(x=monthly_revenue.index, y=decomposition.observed,
                         mode='lines', name='Observed', line=dict(color='royalblue')), row=1, col=1)

fig.add_trace(go.Scatter(x=monthly_revenue.index, y=decomposition.trend,
                         mode='lines', name='Trend', line=dict(color='darkgreen')), row=2, col=1)

fig.add_trace(go.Scatter(x=monthly_revenue.index, y=decomposition.seasonal,
                         mode='lines', name='Seasonality', line=dict(color='orange')), row=3, col=1)

fig.add_trace(go.Scatter(x=monthly_revenue.index, y=decomposition.resid,
                         mode='lines', name='Residuals', line=dict(color='firebrick')), row=4, col=1)

# Step 5: Layout styling
fig.update_layout(
    height=900,
    width=1000,
    title_text='Time Series Decomposition of Monthly Revenue (6-Month Seasonality)',
    title_font=dict(size=20),
    plot_bgcolor='white',
    paper_bgcolor='white',
    margin=dict(t=80, l=40, r=20, b=40)
)

fig.update_xaxes(title_text='Date', row=4, col=1)
fig.update_yaxes(title_text='Revenue')

fig.show()
```

## Detailed Insight – Revenue Structure Breakdown

---

# 20. ARIMA Forecast of Monthly Revenue Trend

This forecast uses the **trend component** from time series decomposition and fits an **ARIMA(1,1,1)** model to project revenue for the next 6 months.

The model captures:
- Long-term movement in revenue (removes seasonality and noise)
- Confidence intervals for planning uncertainty
- Likely short-term fluctuations around the trend

The goal is to support **realistic target-setting** and **resource planning**.


```python
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'decomposition' object is available from the previous cell execution
# Extract the trend component from the decomposition object
trend_series = decomposition.trend.dropna() # Drop potential NaN values at the ends

# Fit ARIMA model on trend
model = ARIMA(trend_series, order=(1, 1, 1))
model_fit = model.fit()

# Forecast next 6 months
forecast_steps = 6
forecast_result = model_fit.get_forecast(steps=forecast_steps)
forecast_values = forecast_result.predicted_mean
forecast_ci = forecast_result.conf_int()

# Prepare forecast index and bounds
forecast_index = pd.date_range(start=trend_series.index[-1] + pd.offsets.MonthBegin(1), periods=forecast_steps, freq='ME')
forecast_values_array = forecast_values.values.astype(float)
lower_bound = forecast_ci.iloc[:, 0].values.astype(float)
upper_bound = forecast_ci.iloc[:, 1].values.astype(float)

# Create Plotly figure
fig = go.Figure()

# Historical trend
fig.add_trace(go.Scatter(
    x=trend_series.index,
    y=trend_series.values,
    mode='lines',
    name='Historical Trend',
    line=dict(color='royalblue', width=3)
))

# Forecast values
fig.add_trace(go.Scatter(
    x=forecast_index,
    y=forecast_values_array,
    mode='lines+markers',
    name='Forecast',
    line=dict(color='darkorange', width=3)
))

# Confidence interval
fig.add_trace(go.Scatter(
    x=np.concatenate([forecast_index, forecast_index[::-1]]),
    y=np.concatenate([upper_bound, lower_bound[::-1]]),
    fill='toself',
    fillcolor='rgba(255, 165, 0, 0.2)',
    line=dict(color='rgba(255,255,255,0)'),
    hoverinfo="skip",
    showlegend=True,
    name='95% Confidence Interval'
))

# Layout styling
fig.update_layout(
    title='ARIMA Forecast of Monthly Revenue Trend (Next 6 Months)',
    title_font=dict(size=20),
    xaxis_title='Date',
    yaxis_title='Revenue (₦)',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(size=12),
    margin=dict(t=80, l=40, r=20, b=40),
    hovermode='x unified'
)

fig.show()
```

## Detailed Insight – Forecast Takeaways

---

### I. Short-Term Stability with Gradual Growth
- The forecast indicates a **modest upward trend** over the next 6 months.
- Aligns with prior trends and suggests that revenue momentum is **stable but unspectacular**.

---

### II. Wide Confidence Bounds Suggest External Sensitivity
- The 95% confidence interval widens slightly over time — reflecting **increased uncertainty** farther into the future.
- Business decisions (example:- staffing, promotions) should account for both upper and lower bounds.

---

### III. Trend vs. Actual Monitoring
- Future revenue should be monitored **against this trend baseline** to flag underperformance or positive breakout.
- Use this forecast as a benchmark to **evaluate H2 campaign impact**.

> ARIMA lets you quantify "what happens next" — and plan with foresight.





# 21. Revenue Trends by Product – H1 2023 vs H1 2024

This analysis compares **first-half revenue performance** (Jan–Jun) across two consecutive years, by product line.

We calculate:

- **Total Revenue** per product per year  
- **Year-over-Year (YoY) Growth** as a %  
- **Z-score** to indicate how extreme each year’s performance is relative to the product’s average

This helps identify top movers, lagging products, and outliers worth deeper exploration.


```python
from scipy.stats import f_oneway, kruskal

# Step 1: Filter data for Jan–Jun 2023 and Jan–Jun 2024
business_performance['date'] = pd.to_datetime(business_performance['date'])
filtered_data = business_performance[
    ((business_performance['date'].dt.year == 2023) & (business_performance['date'].dt.month <= 6)) |
    ((business_performance['date'].dt.year == 2024) & (business_performance['date'].dt.month <= 6))
]

filtered_data['year'] = filtered_data['date'].dt.year

# Step 2: Aggregate revenue by product and year
revenue_yearly = filtered_data.groupby(['product', 'year'])['revenue'].sum().reset_index()

# Step 3: Compute YoY growth and Z-scores
revenue_yearly['YoY_growth'] = revenue_yearly.groupby('product')['revenue'].pct_change()
revenue_yearly['Z_score'] = revenue_yearly.groupby('product')['revenue'].transform(lambda x: (x - x.mean()) / x.std())

# Step 4: Round values and format for display
revenue_table = revenue_yearly.copy()
revenue_table['revenue'] = revenue_table['revenue'].apply(lambda x: f"₦{x:,.0f}")
revenue_table['YoY_growth'] = revenue_table['YoY_growth'].apply(lambda x: f"{x:.1%}" if pd.notnull(x) else "—")
revenue_table['Z_score'] = revenue_table['Z_score'].apply(lambda x: f"{x:.2f}" if pd.notnull(x) else "—")

# Step 5: Create Plotly Table
fig = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>Product</b>", "<b>Year</b>", "<b>Total Revenue</b>", "<b>YoY Growth</b>", "<b>Z-Score</b>"],
        fill_color='#1f2e5a',
        font=dict(color='white', size=14),
        align='left',
        height=35
    ),
    cells=dict(
        values=[
            revenue_table['product'],
            revenue_table['year'],
            revenue_table['revenue'],
            revenue_table['YoY_growth'],
            revenue_table['Z_score']
        ],
        fill_color='aliceblue',
        align='left',
        font=dict(size=13),
        height=30
    )
)])

# Step 6: Layout Styling
fig.update_layout(
    title='Revenue Trends by Product (Jan–Jun 2023 vs Jan–Jun 2024)',
    title_font=dict(size=20),
    height=500,
    margin=dict(l=20, r=20, t=80, b=20)
)

fig.show()
```

## Detailed Insight – H1 Revenue Performance by Product

---

### I. Revenue Growth Stability
- All products recorded **positive revenue** across both years.
- **YoY growth margins are modest** — indicating early stability but no standout surge.
- Suggests retention of base customers rather than strong new acquisition.

---

### II. Z-Score Differentiation
- Z-scores highlight **which revenue values deviate from product averages**.
- A Z-score > +1 or < -1 signals significant over- or underperformance worth root cause analysis.

---

### III. Relative Movers
- Products with moderate revenue but **high YoY growth** are potential up-and-comers.
- Products with **flat growth and low Z-scores** may require strategic review or reactivation.

# 22. Revenue Trend Comparison: Jan–Jun 2023 vs Jan–Jun 2024

This visual comparison of **monthly revenue** between H1 2023 and H1 2024 reveals changes in seasonal performance, campaign effectiveness, and growth momentum.

It helps answer:

- Are we improving over the same period last year?
- Which months are outperforming or underperforming?
- Is the revenue trajectory consistent or volatile?


```python
# Helper function to format revenue
def format_naira(value):
    if value >= 1e9:
        return f"{value / 1e9:.2f}B"
    elif value >= 1e6:
        return f"{value / 1e6:.2f}M"
    else:
        return f"{value:,.0f}"

# Prepare data
monthly_revenue = business_performance.copy()
monthly_revenue['date'] = pd.to_datetime(monthly_revenue['date'])

# Filter only Jan–Jun for 2023 and 2024
monthly_revenue = monthly_revenue[
    ((monthly_revenue['date'].dt.year == 2023) | (monthly_revenue['date'].dt.year == 2024)) &
    (monthly_revenue['date'].dt.month <= 6)
]

# Extract month and year
monthly_revenue['month'] = monthly_revenue['date'].dt.month
monthly_revenue['year'] = monthly_revenue['date'].dt.year

# Aggregate
trend_data = monthly_revenue.groupby(['year', 'month'])['revenue'].sum().reset_index()

# Map month numbers to names for x-axis
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun'}
trend_data['month_name'] = trend_data['month'].map(month_map)
trend_data['month_order'] = trend_data['month']
trend_data = trend_data.sort_values(by='month_order')

# Base line plot
fig = px.line(
    trend_data,
    x='month_name',
    y='revenue',
    color='year',
    markers=True,
    title='Revenue Trend Comparison: Jan–Jun 2023 vs Jan–Jun 2024',
    labels={'revenue': 'Total Revenue (₦)', 'month_name': 'Month', 'year': 'Year'},
    color_discrete_sequence=px.colors.qualitative.Set1
)

# Add formatted black labels to each point
for _, row in trend_data.iterrows():
    fig.add_annotation(
        x=row['month_name'],
        y=row['revenue'],
        text=format_naira(row['revenue']),
        showarrow=False,
        font=dict(color='black', size=11),
        bgcolor='rgba(255,255,255,0.7)',
        xanchor='center',
        yanchor='bottom'
    )

# Style layout
fig.update_layout(
    title_font=dict(size=20),
    font=dict(size=12),
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='x unified',
    legend_title_text='Year',
    margin=dict(t=80, l=40, r=20, b=40)
)

fig.show()
```
