# 1. Monthly Revenue Trend Analysis by Product

This section analyzes how monthly revenue varies across FinTechLend's three main lending products: **Business Loan**, **Personal Loan**, and **BNPL**.

Understanding revenue momentum at the product level helps identify:
- Seasonal patterns or revenue peaks,
- Product performance consistency,
- Opportunities for prioritizing product-specific growth strategies.

We use Plotly to generate an interactive line chart with monthly data aggregated per product.


```python
# Step 1: Convert date and prepare data
business_performance['date'] = pd.to_datetime(business_performance['date'])

# Step 2: Aggregate monthly by product
monthly_perf = business_performance.groupby(
    [pd.Grouper(key='date', freq='ME'), 'product']
).agg({
    'application_volume': 'sum',
    'conversion_rate': 'mean',
    'acquisition_cost': 'mean',
    'revenue': 'sum',
    'margin': 'mean'
}).reset_index()

# Format date for display
monthly_perf['month_label'] = monthly_perf['date'].dt.strftime('%b-%y')
monthly_perf = monthly_perf.sort_values(by='date')

# Step 3: Plot using Seaborn/Matplotlib
plt.figure(figsize=(12, 7))
sns.set(style="whitegrid")

# Plot each product line
for product in monthly_perf['product'].unique():
    product_data = monthly_perf[monthly_perf['product'] == product]
    plt.plot(
        product_data['month_label'],
        product_data['revenue'],
        marker='o',
        label=product
    )

# Step 4: Styling
plt.title('Monthly Revenue Trend by Product', fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue (₦)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Product', fontsize=10, title_fontsize=11)
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show plot
plt.show()
```

# 2. Channel Efficiency Analysis: Acquisition Cost vs Revenue

This section evaluates the performance of FinTechLend's acquisition channels based on three metrics:
- **Average Acquisition Cost (CAC)**  
- **Total Revenue Generated**  
- **Conversion Rate (%)**

We use a **bubble chart** to assess the relative efficiency of each channel — combining cost, return, and conversion size into one visual.

This helps identify high-ROI channels and flag inefficient or overpriced acquisition sources.


```python

# Helper function to format currency
def format_naira(value):
    if value >= 1e12:
        return f"₦{value / 1e12:.1f}T"
    elif value >= 1e9:
        return f"₦{value / 1e9:.1f}B"
    elif value >= 1e6:
        return f"₦{value / 1e6:.0f}M"
    else:
        return f"₦{value:,.0f}"

# Step 1: Group and aggregate
channel_perf = business_performance.groupby('channel').agg({
    'application_volume': 'sum',
    'conversion_rate': 'mean',
    'acquisition_cost': 'mean',
    'revenue': 'sum',
    'margin': 'mean'
}).reset_index()

channel_perf = channel_perf.sort_values(by='conversion_rate', ascending=False)

# Normalize bubble sizes for Matplotlib
channel_perf['size_scaled'] = (channel_perf['conversion_rate'] - channel_perf['conversion_rate'].min()) / \
                              (channel_perf['conversion_rate'].max() - channel_perf['conversion_rate'].min()) * 1000 + 100

# Step 2: Plot
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

scatter = plt.scatter(
    x=channel_perf['acquisition_cost'],
    y=channel_perf['revenue'],
    s=channel_perf['size_scaled'],
    c=range(len(channel_perf)),
    cmap='Set1',
    edgecolor='black',
    alpha=0.7
)

# Step 3: Add annotations
for i, row in channel_perf.iterrows():
    label = f"{row['channel']}\nCost: {format_naira(row['acquisition_cost'])}\nRev: {format_naira(row['revenue'])}"
    plt.text(
        row['acquisition_cost'],
        row['revenue'],
        label,
        fontsize=9,
        ha='center',
        va='center',
        bbox=dict(facecolor='white', edgecolor='gray', alpha=0.5, boxstyle='round,pad=0.3')
    )

# Step 4: Styling
plt.title('Channel Efficiency: Acquisition Cost vs Revenue', fontsize=18, fontweight='bold')
plt.xlabel('Avg Acquisition Cost (₦)', fontsize=12)
plt.ylabel('Total Revenue (₦)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show
plt.show()
```

## Detailed Insight – Channel Efficiency (CAC vs Revenue)

---

### I. Partner Channel – High Revenue, High Cost
- Generates the **highest total revenue** (₦116M) and leads in conversion rate (8%).
- However, it comes at a premium, with the **highest acquisition cost (₦300)**.
- Best suited for high-intent, high-value acquisition bursts — not ideal for always-on scaling.

---

### II. Direct Channel – Most Efficient
- Achieves a strong **revenue output (₦72M)** at the **lowest CAC (₦50)** with a solid 5% conversion rate.
- Represents the **most scalable and cost-efficient channel**.
- Should be prioritized in both paid and organic growth campaigns.

---

### III. Content & Social Media – Low ROI
- Both channels show **low revenue (₦29M–₦43M)** despite moderate acquisition costs (₦100–₦150).
- Conversion rates are the **lowest (2–3%)**, indicating a misalignment between message and market.
- These should be **re-evaluated or repositioned** for top-of-funnel rather than direct acquisition.

---

### Strategic Takeaways
- **Double down on Direct** for efficient scaling.
- **Use Partner for spikes**, but manage CAC carefully.
- **Audit low-performing channels** (example:- Social and Content) for creative, audience, or channel mismatch.
- Consider introducing **multi-touch attribution** to better value early-stage influence channels.



# 3. Customer Sentiment by Segment – Satisfaction vs NPS

Understanding how different customer segments perceive FinTechLend's service is crucial to reducing churn and improving loyalty. This analysis maps each customer segment by:

- **Satisfaction Score (1–5 scale)**
- **Net Promoter Score (NPS)**
- **Feedback Volume**

We use a **bubble chart** to highlight high-impact segments that may require targeted retention, better onboarding, or tailored support improvements.


```python

# Convert date column
customer_feedback['date'] = pd.to_datetime(customer_feedback['date'])

# Step 1: Aggregate feedback
segment_feedback = customer_feedback.groupby('customer_segment').agg({
    'nps_score': 'mean',
    'satisfaction_score': 'mean',
    'feedback_volume': 'sum'
}).reset_index()

# Helper to format feedback volume
def format_volume(v):
    if v >= 1e6:
        return f"{v / 1e6:.1f}M"
    elif v >= 1e3:
        return f"{v / 1e3:.0f}K"
    else:
        return str(int(v))

# Step 2: Normalize bubble size for visibility
segment_feedback['size_scaled'] = (
    (segment_feedback['feedback_volume'] - segment_feedback['feedback_volume'].min()) /
    (segment_feedback['feedback_volume'].max() - segment_feedback['feedback_volume'].min()) * 1000 + 100
)

# Step 3: Plot
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Bubble chart
scatter = plt.scatter(
    x=segment_feedback['satisfaction_score'],
    y=segment_feedback['nps_score'],
    s=segment_feedback['size_scaled'],
    c=range(len(segment_feedback)),
    cmap='tab10',
    edgecolor='black',
    alpha=0.75
)

# Step 4: Add labels inside bubbles
for i, row in segment_feedback.iterrows():
    label = f"{row['customer_segment']}\nS:{row['satisfaction_score']:.1f} N:{row['nps_score']:.1f}\nF:{format_volume(row['feedback_volume'])}"
    plt.text(
        row['satisfaction_score'],
        row['nps_score'],
        label,
        fontsize=9,
        ha='center',
        va='center',
        bbox=dict(facecolor='white', edgecolor='gray', alpha=0.5, boxstyle='round,pad=0.3')
    )

# Step 5: Final styling
plt.title('Customer Sentiment by Segment: Satisfaction vs NPS', fontsize=18, fontweight='bold')
plt.xlabel('Avg Satisfaction Score', fontsize=12)
plt.ylabel('Avg NPS Score', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show plot
plt.show()
```

## Detailed Insight – Customer Sentiment by Segment

---

### I. High-Value Advocates (Gen Z, Small Business)
- These segments show **strong satisfaction (4.5+) and high NPS (>45)**.
- Feedback volume is moderate but influential, suggesting positive user experience and referral potential.
- Opportunity: Use these groups to drive word-of-mouth campaigns and referral incentives.

---

### II. At-Risk Segments (Mid-Career, Startups)
- These segments have **lower-than-average NPS (44.5)** and moderate satisfaction (4.3–4.4).
- Feedback volumes are high, implying **engaged but dissatisfied users**.
- Priority for churn prevention and service recovery efforts.

---

### III. Stable but Passive (Gen X, Medium Enterprises)
- These users sit near the average on both axes (Satisfaction 4.4, NPS 45.0).
- Their experience is neither negative nor enthusiastic — a potential **“silent churn”** group.
- Opportunity: Re-engage through onboarding refresh, surveys, or proactive support.

---

### Strategic Takeaways
- **Protect and amplify your promoters** (Gen Z, Small Business) through loyalty, recognition, and referral initiatives.
- **Fix friction points** for Mid-Career and Startup users with faster feedback loops and targeted CX improvements.
- **Reinvigorate the middle** — Gen X and Medium Enterprises — with personalized journeys and engagement nudges.

# 5. Market Size vs Competitor Revenue Over Time

This analysis tracks the evolution of the total addressable market and competitor revenue in the fintech lending space. It helps us understand:

- How fast the market is growing (₦ trillions per month)
- Whether competitors are scaling proportionately
- Where FinTechLend sits in this competitive trajectory

The dual-line chart compares **Market Size** and **Competitor Revenue** monthly, offering a clear view of scale and share shifts.


```python

# Format large numbers
def format_naira(value):
    if value >= 1e12:
        return f"₦{value / 1e12:.1f}T"
    elif value >= 1e9:
        return f"₦{value / 1e9:.0f}B"
    elif value >= 1e6:
        return f"₦{value / 1e6:.0f}M"
    else:
        return f"₦{value:,.0f}"

# Clean and prepare data
market_data.columns = market_data.columns.str.strip()
market_data['market_size'] = market_data['market_size'].replace('[\$,]', '', regex=True).astype(float)
market_data['competitor_revenue'] = market_data['competitor_revenue'].replace('[\$,]', '', regex=True).astype(float)
market_data['date'] = pd.to_datetime(market_data['date'])

# Aggregate
unique_market_sizes = market_data.drop_duplicates(subset=['date', 'product'])[['date', 'market_size']]
monthly_market_size = unique_market_sizes.groupby(pd.Grouper(key='date', freq='ME'))['market_size'].sum().reset_index()
monthly_revenue = market_data.groupby(pd.Grouper(key='date', freq='ME'))['competitor_revenue'].sum().reset_index()

# Merge and format
market_trend = pd.merge(monthly_market_size, monthly_revenue, on='date')
market_trend['month_label'] = market_trend['date'].dt.strftime('%b-%y')

# Plot
fig, ax1 = plt.subplots(figsize=(14, 7))
ax2 = ax1.twinx()

# Line 1: Market Size
ax1.plot(
    market_trend['month_label'],
    market_trend['market_size'],
    label='Market Size (₦)',
    color='steelblue',
    linewidth=2,
    marker='o'
)

# Line 2: Competitor Revenue
ax2.plot(
    market_trend['month_label'],
    market_trend['competitor_revenue'],
    label='Competitor Revenue (₦)',
    color='firebrick',
    linewidth=2,
    marker='s'
)

# Add annotations
for i, row in market_trend.iterrows():
    ax1.annotate(
        format_naira(row['market_size']),
        (row['month_label'], row['market_size']),
        textcoords="offset points",
        xytext=(0, 5),
        ha='center',
        fontsize=9,
        color='black',
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray')
    )
    ax2.annotate(
        format_naira(row['competitor_revenue']),
        (row['month_label'], row['competitor_revenue']),
        textcoords="offset points",
        xytext=(0, -15),
        ha='center',
        fontsize=9,
        color='black',
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray')
    )

# Axis styling
ax1.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Market Size (₦)', color='steelblue', fontsize=12)
ax2.set_ylabel('Competitor Revenue (₦)', color='firebrick', fontsize=12)
ax1.tick_params(axis='y', labelcolor='steelblue')
ax2.tick_params(axis='y', labelcolor='firebrick')
ax1.tick_params(axis='x', rotation=45)

# Title and grid
plt.title('Market Size vs Competitor Revenue Over Time', fontsize=18, fontweight='bold', pad=20)
ax1.grid(True, which='major', linestyle='--', linewidth=0.5)
fig.tight_layout()

# Legend
fig.legend(
    loc='upper left',
    bbox_to_anchor=(0.1, 0.95),
    bbox_transform=fig.transFigure,
    frameon=True
)

# Show plot
plt.show()
```

# 6. FinTechLend Monthly Revenue & Market Share

This analysis computes and visualizes FinTechLend's share of the total digital lending market each month. It integrates internal revenue data with external market data to highlight:

- Monthly revenue growth or stagnation  
- Total market expansion  
- Market share positioning (relative to competitors)

A tabular summary provides a clear, accessible breakdown of how well FinTechLend is keeping pace with industry growth.


```python

# Helper functions
def format_naira(val):
    if val >= 1e12: return f"₦{val / 1e12:.2f}T"
    elif val >= 1e9: return f"₦{val / 1e9:.0f}B"
    elif val >= 1e6: return f"₦{val / 1e6:.0f}M"
    else: return f"₦{val:,.0f}"

def format_arrow(change):
    if pd.isnull(change): return ''
    return '↑' if change > 0 else ('↓' if change < 0 else '')

def format_share(value, change):
    if pd.isnull(value): return ''
    percent = f"{value * 100:.1f}%"
    arrow = format_arrow(change)
    return f"{percent} {arrow}"

# Prep the data
business_performance['date'] = pd.to_datetime(business_performance['date'])
market_data['date'] = pd.to_datetime(market_data['date'])
market_data['market_size'] = market_data['market_size'].replace('[\$,]', '', regex=True).astype(float)
market_data['competitor_revenue'] = market_data['competitor_revenue'].replace('[\$,]', '', regex=True).astype(float)

fintechlend_monthly = business_performance.groupby(pd.Grouper(key='date', freq='ME')).agg({
    'revenue': 'sum'
}).rename(columns={'revenue': 'FinTechLend_Revenue'}).reset_index()

market_monthly = market_data.groupby(pd.Grouper(key='date', freq='ME')).agg({
    'market_size': 'sum',
    'competitor_revenue': 'sum'
}).reset_index()

merged = pd.merge(fintechlend_monthly, market_monthly, on='date', how='left')
merged['Total_Market_Size'] = merged['FinTechLend_Revenue'] + merged['competitor_revenue']
merged['FinTechLend_Market_Share'] = merged['FinTechLend_Revenue'] / merged['Total_Market_Size']
merged['Market_Share_Change'] = merged['FinTechLend_Market_Share'].diff()
merged['Month'] = merged['date'].dt.strftime('%b-%Y')

# Format for table
merged['Revenue'] = merged['FinTechLend_Revenue'].apply(format_naira)
merged['Market Size'] = merged['Total_Market_Size'].apply(format_naira)
merged['Market Share'] = merged.apply(lambda r: format_share(r['FinTechLend_Market_Share'], r['Market_Share_Change']), axis=1)

table_data = merged[['Month', 'Revenue', 'Market Size', 'Market Share']].copy()
table_data.columns = ['Month', 'Revenue', 'Market Size', 'Market Share']

# Create the figure and table
fig, ax = plt.subplots(figsize=(10, 0.7 * len(table_data) + 2))
ax.axis('off')

# Add table
mpl_table = plt.table(
    cellText=table_data.values,
    colLabels=table_data.columns,
    cellLoc='left',
    loc='center',
    colColours=['#003366']*4
)

# Style table
mpl_table.auto_set_font_size(False)
mpl_table.set_fontsize(12)

# Header style
for k, cell in mpl_table.get_celld().items():
    if k[0] == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_fontsize(13)
    cell.set_edgecolor('lightgray')
    if k[0] > 0:
        cell.set_facecolor('#f7f7f7' if k[0] % 2 == 0 else '#ffffff')

# Add title
plt.title('FinTechLend Monthly Revenue & Market Share', fontsize=16, weight='bold', pad=20)

plt.tight_layout()
plt.show()


```

# 17. Group Performance Test – Conversion Rate by Channel

This test checks whether **conversion rates differ significantly across acquisition channels** using two statistical methods:

- **One-Way ANOVA** – assumes normal distribution and equal variances  
- **Kruskal-Wallis H Test** – a non-parametric alternative, does not assume normality

These methods help confirm whether observed differences in conversion rate are **statistically significant** or due to random variation.


```python
from scipy.stats import f_oneway, kruskal

# Step 1: Group conversion rates by channel
channel_groups = [
    group['conversion_rate'].values
    for _, group in business_performance.groupby('channel')
]

# Step 2: Run tests
anova_result = f_oneway(*channel_groups)
kruskal_result = kruskal(*channel_groups)

# Step 3: Prepare summary
anova_summary = {
    "Metric": [
        "ANOVA F-statistic",
        "ANOVA p-value",
        "Kruskal-Wallis H-statistic",
        "Kruskal-Wallis p-value"
    ],
    "Value": [
        round(anova_result.statistic, 4),
        round(anova_result.pvalue, 4),
        round(kruskal_result.statistic, 4),
        round(kruskal_result.pvalue, 4)
    ]
}
summary_df = pd.DataFrame(anova_summary)

# Step 4: Plot table
fig, ax = plt.subplots(figsize=(8, 2))
ax.axis('off')

table = plt.table(
    cellText=summary_df.values,
    colLabels=summary_df.columns,
    cellLoc='left',
    loc='center',
    colColours=['#003366', '#003366']
)

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(12)

for (row, col), cell in table.get_celld().items():
    cell.set_edgecolor('gray')
    if row == 0:
        cell.set_text_props(weight='bold', color='white')  # header row
    else:
        cell.set_facecolor('#f4f4f4' if row % 2 == 0 else '#ffffff')

# Title
plt.title('Group Performance Statistical Test Summary', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

## Detailed Insight – Are Channel Conversion Rates Statistically Different?

---

### ANOVA Result
- **F-statistic** indicates how much variance exists between group means relative to within-group variance.
- A **p-value < 0.05** would mean **at least one channel** differs significantly in conversion rate.
- If p > 0.05, we fail to reject the null hypothesis that all means are equal.

---

### Kruskal-Wallis Result
- Non-parametric alternative to ANOVA (useful when distributions aren't normal).
- Similar interpretation: **low p-value means significant differences** between at least one pair of groups.

---

### Interpretation Guidelines

- If **both p-values < 0.05** → Significant differences exist across channels. Proceed with post-hoc tests (e.g., Tukey HSD or Dunn's test).
- If **both p-values ≥ 0.05** → No statistically significant difference in conversion rate by channel. Focus on operational improvements.

---

### Strategic Implication

- **Statistically significant results** justify **channel-specific optimization** or investment.
- If not significant, observed conversion differences may not be meaningful — prioritize **cost efficiency and volume** instead.

> This test ensures data-driven decisions when comparing performance across acquisition channels.



# 18. Churn Prediction Model: Feature Importance

This analysis builds a **logistic regression model** to predict churn likelihood for customers with **satisfaction scores < 3**.

We analyze the effect of:
- **NPS score**
- **Feedback volume**
- **Product type (one-hot encoded)**

The resulting coefficients explain which variables are **positively or negatively associated with churn**, helping inform targeted retention strategies.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Step 1: Define churn as satisfaction_score < 3
customer_feedback['churn_flag'] = (customer_feedback['satisfaction_score'] < 3).astype(int)

# Step 2: Balance classes
churn_data = customer_feedback[customer_feedback['churn_flag'] == 1].sample(n=100, replace=True, random_state=1)
non_churn_data = customer_feedback[customer_feedback['churn_flag'] == 0].sample(n=100, replace=True, random_state=1)
churn_model_data = pd.concat([churn_data, non_churn_data])

# Step 3: Prepare features and target
features = pd.get_dummies(churn_model_data[['nps_score', 'feedback_volume', 'product']], drop_first=True)
target = churn_model_data['churn_flag']

# Step 4: Train-test split and scale
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 5: Train model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_scaled, y_train)

# Step 6: Extract coefficients
coef_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': logreg.coef_[0]
}).sort_values(by='Coefficient', ascending=True)

# Step 7: Plot coefficients
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')

barplot = sns.barplot(
    data=coef_df,
    x='Coefficient',
    y='Feature',
    palette='RdBu_r'
)

# Add coefficient labels inside bars
for index, row in coef_df.iterrows():
    plt.text(
        row['Coefficient'] + 0.02 if row['Coefficient'] >= 0 else row['Coefficient'] - 0.08,
        index,
        f"{row['Coefficient']:.2f}",
        color='white' if abs(row['Coefficient']) > 0.15 else 'black',
        fontsize=10,
        ha='left' if row['Coefficient'] >= 0 else 'right',
        va='center',
        bbox=dict(facecolor='black' if abs(row['Coefficient']) > 0.15 else 'none', alpha=0.6 if abs(row['Coefficient']) > 0.15 else 0)
    )

# Final touches
plt.title('Churn Predictor Importance (Satisfaction Score < 3)', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Logistic Regression Coefficient')
plt.ylabel('Predictor')
plt.tight_layout()
plt.show()
```

# 19. Time Series Decomposition of Monthly Revenue

This analysis applies an **additive decomposition** to monthly revenue using a **6-month seasonality window**. The goal is to isolate:

- **Trend** – long-term direction of revenue  
- **Seasonality** – repeated cyclical patterns (example:- biannual spikes)  
- **Residuals** – unexplained variation or noise

This breakdown helps diagnose whether revenue changes are structural, cyclical, or random.


```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 1: Prepare time series data
ts_data = business_performance.copy()
ts_data['date'] = pd.to_datetime(ts_data['date'])

monthly_revenue = ts_data.groupby(pd.Grouper(key='date', freq='ME'))['revenue'].sum()

# Step 2: Decompose (Additive Model, 6-month seasonality)
decomposition = seasonal_decompose(monthly_revenue, model='additive', period=6)

# Step 3: Plot using Matplotlib
fig, axes = plt.subplots(4, 1, figsize=(12, 10), sharex=True)
fig.suptitle('Time Series Decomposition of Monthly Revenue (6-Month Seasonality)', fontsize=18, fontweight='bold', y=1.02)

# Observed
axes[0].plot(decomposition.observed, color='royalblue')
axes[0].set_title('Observed Revenue', fontsize=14)
axes[0].grid(True, linestyle='--', linewidth=0.5)

# Trend
axes[1].plot(decomposition.trend, color='darkgreen')
axes[1].set_title('Trend Component', fontsize=14)
axes[1].grid(True, linestyle='--', linewidth=0.5)

# Seasonal
axes[2].plot(decomposition.seasonal, color='orange')
axes[2].set_title('Seasonal Component', fontsize=14)
axes[2].grid(True, linestyle='--', linewidth=0.5)

# Residual
axes[3].plot(decomposition.resid, color='firebrick')
axes[3].set_title('Residuals', fontsize=14)
axes[3].set_xlabel('Date')
axes[3].grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

## Detailed Insight – Revenue Structure Breakdown

---

### I. Trend Component
- Shows the **underlying growth trajectory** — relatively flat or slightly upward.
- Suggests revenue is stable but not accelerating.
- Strategic implication: growth needs **activation via campaigns**, not inertia.

---

### II. Seasonal Pattern
- Reveals **regular peaks and dips every 6 months**, potentially tied to marketing cycles or external factors.
- Aligns with biannual product updates or market rhythms.
- Opportunity: time campaigns to amplify natural spikes.

---

### III. Residuals
- Spikes in residuals suggest **unexpected volatility** or one-off impacts.
- Large residuals should trigger root cause analysis (example:- service outage, competitor action and pricing test).

---

### Strategic Takeaways

- Investigate and explain outliers in residuals to improve predictability.

# 20. ARIMA Forecast of Monthly Revenue Trend

This forecast uses the **trend component** from time series decomposition and fits an **ARIMA(1,1,1)** model to project revenue for the next 6 months.

The model captures:
- Long-term movement in revenue (removes seasonality and noise)
- Confidence intervals for planning uncertainty
- Likely short-term fluctuations around the trend

The goal is to support **realistic target-setting** and **resource planning**.


```python
from statsmodels.tsa.arima.model import ARIMA

# Use the trend component from seasonal decomposition
trend_series = decomposition.trend.dropna()

# Fit ARIMA model
model = ARIMA(trend_series, order=(1, 1, 1))
model_fit = model.fit()

# Forecast next 6 months
forecast_steps = 6
forecast_result = model_fit.get_forecast(steps=forecast_steps)
forecast_values = forecast_result.predicted_mean
forecast_ci = forecast_result.conf_int()

# Create forecast index
forecast_index = pd.date_range(
    start=trend_series.index[-1] + pd.offsets.MonthBegin(1),
    periods=forecast_steps,
    freq='ME'
)

# Extract bounds
lower_bound = forecast_ci.iloc[:, 0].values
upper_bound = forecast_ci.iloc[:, 1].values

# Step 1: Plot
plt.figure(figsize=(12, 6))
plt.plot(trend_series.index, trend_series.values, label='Historical Trend', color='royalblue', linewidth=2)
plt.plot(forecast_index, forecast_values, label='Forecast', color='darkorange', marker='o', linewidth=2)

# Step 2: Fill confidence interval
plt.fill_between(
    forecast_index,
    lower_bound,
    upper_bound,
    color='orange',
    alpha=0.2,
    label='95% Confidence Interval'
)

# Step 3: Styling
plt.title('ARIMA Forecast of Monthly Revenue Trend (Next 6 Months)', fontsize=16, weight='bold', pad=15)
plt.xlabel('Date')
plt.ylabel('Revenue (₦)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(loc='upper left')
plt.tight_layout()

# Step 4: Show
plt.show()
```

## Detailed Insight – Forecast Takeaways

---

### I. Short-Term Stability with Gradual Growth
- The forecast indicates a **modest upward trend** over the next 6 months.
- Aligns with prior trends and suggests that revenue momentum is **stable but unspectacular**.

---

### II. Wide Confidence Bounds Suggest External Sensitivity
- The 95% confidence interval widens slightly over time — reflecting **increased uncertainty** farther into the future.
- Business decisions (example:- staffing, promotions) should account for both upper and lower bounds.

---

### III. Trend vs. Actual Monitoring
- Future revenue should be monitored **against this trend baseline** to flag underperformance or positive breakout.

We calculate:

- **Total Revenue** per product per year  
- **Year-over-Year (YoY) Growth** as a %  
- **Z-score** to indicate how extreme each year’s performance is relative to the product’s average

This helps identify top movers, lagging products, and outliers worth deeper exploration.


```python
# Step 1: Filter Jan–Jun 2023 and 2024
business_performance['date'] = pd.to_datetime(business_performance['date'])
filtered_data = business_performance[
    ((business_performance['date'].dt.year == 2023) & (business_performance['date'].dt.month <= 6)) |
    ((business_performance['date'].dt.year == 2024) & (business_performance['date'].dt.month <= 6))
].copy()

filtered_data['year'] = filtered_data['date'].dt.year

# Step 2: Aggregate revenue
revenue_yearly = filtered_data.groupby(['product', 'year'])['revenue'].sum().reset_index()

# Step 3: YoY Growth and Z-Score
revenue_yearly['YoY_growth'] = revenue_yearly.groupby('product')['revenue'].pct_change()
revenue_yearly['Z_score'] = revenue_yearly.groupby('product')['revenue'].transform(lambda x: (x - x.mean()) / x.std())

# Step 4: Format for table
revenue_table = revenue_yearly.copy()
revenue_table['Total Revenue'] = revenue_table['revenue'].apply(lambda x: f"₦{x:,.0f}")
revenue_table['YoY Growth'] = revenue_table['YoY_growth'].apply(lambda x: f"{x:.1%}" if pd.notnull(x) else "—")
revenue_table['Z-Score'] = revenue_table['Z_score'].apply(lambda x: f"{x:.2f}" if pd.notnull(x) else "—")
revenue_table = revenue_table[['product', 'year', 'Total Revenue', 'YoY Growth', 'Z-Score']]
revenue_table.columns = ['Product', 'Year', 'Total Revenue', 'YoY Growth', 'Z-Score']

# Step 5: Plot as table
fig, ax = plt.subplots(figsize=(10, 0.6 * len(revenue_table) + 2))
ax.axis('off')

table = plt.table(
    cellText=revenue_table.values,
    colLabels=revenue_table.columns,
    cellLoc='left',
    loc='center',
    colColours=['#1f2e5a'] * len(revenue_table.columns)
)

# Style table
table.auto_set_font_size(False)
table.set_fontsize(12)

for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', color='white')
    else:
        cell.set_facecolor('#f0f4ff' if row % 2 == 0 else '#ffffff')
    cell.set_edgecolor('lightgray')

# Step 6: Title
plt.title('Revenue Trends by Product (Jan–Jun 2023 vs Jan–Jun 2024)', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

## Detailed Insight – H1 Revenue Performance by Product

---

### I. Revenue Growth Stability
- All products recorded **positive revenue** across both years.
- **YoY growth margins are modest** — indicating early stability but no standout surge.
- Suggests retention of base customers rather than strong new acquisition.

---

### II. Z-Score Differentiation
- Z-scores highlight **which revenue values deviate from product averages**.
- A Z-score > +1 or < -1 signals significant over- or underperformance worth root cause analysis.

---

### III. Relative Movers
- Products with moderate revenue but **high YoY growth** are potential up-and-comers.
- Products with **flat growth and low Z-scores** may require strategic review or reactivation.

### Strategic Takeaways

# 22. Revenue Trend Comparison: Jan–Jun 2023 vs Jan–Jun 2024

This visual comparison of **monthly revenue** between H1 2023 and H1 2024 reveals changes in seasonal performance, campaign effectiveness, and growth momentum.

It helps answer:

- Are we improving over the same period last year?
- Which months are outperforming or underperforming?
- Is the revenue trajectory consistent or volatile?


```python
# === Data prep ===
df = business_performance.copy()
df['date'] = pd.to_datetime(df['date'])
df = df[
    ((df['date'].dt.year == 2023) | (df['date'].dt.year == 2024)) &
    (df['date'].dt.month <= 6)
].copy()

df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# Aggregate monthly revenue
trend_data = df.groupby(['year', 'month'])['revenue'].sum().reset_index()

# Map month number to name
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun'}
trend_data['month_name'] = trend_data['month'].map(month_map)
trend_data['month_order'] = trend_data['month']

# Sort for correct plot order
trend_data = trend_data.sort_values(by=['year', 'month_order'])

# Helper: Format Naira
def format_naira(value):
    if value >= 1e9:
        return f"₦{value / 1e9:.2f}B"
    elif value >= 1e6:
        return f"₦{value / 1e6:.2f}M"
    else:
        return f"₦{value:,.0f}"

# === Plot ===
fig, ax = plt.subplots(figsize=(10, 6))

# Line plot for each year
for yr in trend_data['year'].unique():
    subset = trend_data[trend_data['year'] == yr]
    ax.plot(
        subset['month_name'],
        subset['revenue'],
        marker='o',
        label=f'{yr}'
    )

    # Add annotation for each point
    for i, row in subset.iterrows():
        ax.annotate(
            format_naira(row['revenue']),
            (row['month_name'], row['revenue']),
            textcoords="offset points",
            xytext=(0, 8),
            ha='center',
            fontsize=9,
            color='black'
        )

# Style
ax.set_title('Revenue Trend Comparison: Jan–Jun 2023 vs Jan–Jun 2024', fontsize=16, fontweight='bold')
ax.set_xlabel('Month')
ax.set_ylabel('Total Revenue (₦)')
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_facecolor('white')
fig.patch.set_facecolor('white')
ax.legend(title='Year')

plt.tight_layout()
plt.show()
```

## Insight – Jan–Jun Revenue Comparison

---

### I. Early 2024 Shows Consistency but Not Acceleration
- Revenue growth in **Jan–Mar 2024** is flat relative to 2023, suggesting **stability but not significant gain**.
- Growth levers deployed in Q2 2024 may not be yielding expected results yet.

---

### II. Performance Diverges in April–June
- Notable **divergence emerges in May and June**, with either positive or negative separation between years.
- Suggests potential campaign, external, or operational factors influencing late-quarter performance.

---

### III. Month-by-Month Labeling Supports Budget Adjustments
- Clear monthly values aid **forecast validation** and help **spot revenue anomalies** (e.g., sharp dips or jumps).
- Useful for budget pacing and tracking recovery or acceleration.

---

### Strategic Use Cases

- Use this comparison in **board or investor presentations** to show trend vs. last year  
- Support **quarterly performance reviews** with visual monthly breakdowns  
- Drive **target-setting and marketing optimization** for H2
